```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('dataset.csv')
```

```python
df.head(20)
```

```python




def mood_mapper(valence, energy, danceability, tempo, loudness):
    # Happy Energetic (Party) - very upbeat
    if valence >=0.6 and energy >=0.6 and danceability >=0.6 and tempo >=110 and loudness >=-6:
        return "Happy / Energetic (High Beat Party)"
    
    # Happy Energetic (Pleasant) - high valence but softer
    elif valence >=0.6 and energy >=0.5 and -12 <= loudness <= -6:
        return "Happy / Energetic (Pleasant)"
    
    # Uplifting Positive (medium-high tempo & mid-high energy)
    elif 0.4 <= valence <= 0.7 and 0.4 <= energy <= 0.7 and 90 <= tempo <= 120:
        return "Uplifting / Positive"
    
    # Calm / Peaceful (low energy + slow tempo + soft loudness)
    elif valence >= 0.5 and energy <= 0.4 and danceability <= 0.5 and tempo <= 90 and loudness <= -12:
        return "Calm / Peaceful"
    
    # Relaxation / Safety (neutral valence, low energy, soft loudness)
    elif 0.3 <= valence <= 0.6 and energy <= 0.4 and loudness <= -12:
        return "Relaxation / Safety"
    
    # Balanced / Neutral
    elif 0.4 <= valence <= 0.6 and 0.4 <= energy <= 0.6:
        return "Balanced / Neutral"
    
    # Fallback → classify by dominant feature
    else:
        if energy <= 0.3:
            return "Low Energy / Calm"
        elif valence <= 0.3:
            return "Sad / Negative"
        elif energy >= 0.7 and tempo >= 120:
            return "High Energy / Excited"
        else:
            return "Mixed / Uncategorized"


# Apply function
df["MoodCategory"] = df.apply(
    lambda row: mood_mapper(row["valence"], row["energy"], row["danceability"], row["tempo"], row["loudness"]),
    axis=1
)

# Save updated dataset
df.to_csv("dataset_with_moods_full.csv", index=False)

# Show preview
print(df[["track_name", "valence", "energy", "danceability", "tempo", "loudness", "MoodCategory"]].head())
```

```python
merge_map = {
    "Low Energy / Calm": "Calm & Relaxed",
    "Calm / Peaceful": "Calm & Relaxed",
    "Relaxation / Safety": "Calm & Relaxed"
}

# Apply mapping
df["MoodCategory"] = df["MoodCategory"].replace(merge_map)

# Check updated counts
```
