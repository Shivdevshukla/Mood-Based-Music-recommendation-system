```python
import pandas as pd

# Load your dataset
df = pd.read_csv("muse_dataset.csv")

# Function to map VAD values into recommendations
def map_recommendation(valence, arousal, dominance):
    if valence > 0.6 and arousal > 0.6:
        return "Happy/Energetic"
    elif valence > 0.6 and 0.4 <= arousal <= 0.7:
        return "Uplifting/Positive"
    elif valence > 0.5 and arousal < 0.4:
        return "Soothing/Peaceful"
    elif valence < 0.5 and arousal < 0.4:
        return "Relaxation/Safety"
    elif 0.4 <= valence <= 0.6 and 0.4 <= arousal <= 0.6:
        return "Balanced/Moderate"
    else:
        return "Uncategorized"


# Apply function to create new column
df["Recommendation"] = df.apply(lambda row: map_recommendation(row["valence_tags"], row["arousal_tags"],row["dominance_tags"]), axis=1)

# Save new dataset with the added column
df.to_csv("muse_dataset_with_recommendations.csv", index=False)

print(df.head())
```

```python
import pandas as pd

df = pd.read_csv("muse_dataset.csv")

def categorize(value, q1, q3):
    if value <= q1:
        return "Low"
    elif value <= q3:
        return "Medium"
    else:
        return "High"

q_val = df['valence_tags'].quantile([0.25, 0.75])
q_aro = df['arousal_tags'].quantile([0.25, 0.75])
q_dom = df['dominance_tags'].quantile([0.25, 0.75])


df['Valence_Level'] = df['valence_tags'].apply(lambda x: categorize(x, q_val[0.25], q_val[0.75]))
df['Arousal_Level'] = df['arousal_tags'].apply(lambda x: categorize(x, q_aro[0.25], q_aro[0.75]))
df['Dominance_Level'] = df['dominance_tags'].apply(lambda x: categorize(x, q_dom[0.25], q_dom[0.75]))

# Save new dataset
```
